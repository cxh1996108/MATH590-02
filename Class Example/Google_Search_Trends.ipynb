```python
#pip intsall pytrends package, see link https://pypi.org/project/pytrends/#description

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pytrends.request import TrendReq

pytrend = TrendReq()

pytrend.build_payload(kw_list=['silver price'])

# Interest by Region
df = pytrend.interest_by_region()
df = pytrend.interest_over_time()

#df.sort_values (by ='Taylor Swift', ascending= False)

df.plot(figsize = [15,7], marker ="*")
```

```python
df.reset_index().plot(x='geoName', y='Taylor Swift', figsize=(120, 30), kind ='bar')
```

```python
df = pytrend.trending_searches(pn='united_states')
```

```python
# Get Google Keyword Suggestions
keywords = pytrend.suggestions(keyword='Trump')
df = pd.DataFrame(keywords)
df.drop(columns= 'mid')   # This column makes no sense
```

```python
#Related Search

pytrend.build_payload(kw_list=['Coronavirus'])
# Related Queries, returns a dictionary of dataframes
related_queries = pytrend.related_queries()
related_queries.values()
```

```python
# Related Topics, returns a dictionary of dataframes
related_topic = pytrend.related_topics()
related_topic.values()
```
